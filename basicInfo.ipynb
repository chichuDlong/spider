#### 1、爬虫应用分类：
>--通用爬虫：抓取系统重要组成部分。抓取的是一整张页面数据。  
>--聚焦爬虫：建立在通用爬虫基础上。抓取页面汇总特定的局部内容。  
>--增量式爬虫：检测网站中数据更新情况。仅提取网站中最新更新的数据内容。
#### 2、反爬机制
门户网站通过相关策略或技术手段防止爬虫程序进行网站数据爬取。
#### 3、反反爬策略
通过相关策略或技术手段破解反爬机制，进而获得门户网站资源。
#### 4、robots.txt协议
查看网站允许爬取的内容：网站地址+/robots.txt，如：https://translate.google.cn/robots.txt
#### 5、http协议
http(hyper text transfer protocol)：超文本传输协议缩写，表示服务器与客户端进行数据交互的传送协议。  
常用请求头信息(Request Headers)：  
- User-Agent：请求载体的身份标识。  
- Connection：请求完毕后，是断开连接还是保持连接。  
常用响应头信息：  
- Content-Type：服务器响应会客户端的数据类型，通常有字符串、html、json、···
#### 6、https协议
- 安全的超文本传输协议，经过加密的数据传输
#### 7、加密方式
- 1）、对称秘钥加密：客户端将加密的数据与解密的秘钥一同发送至客户端，传输过程中容易被拦截，存在安全隐患；  
- 2）、非对称秘钥加密：服务器端指定好加密方式，客户端根据加密方式（公钥）对数据加密后发送至服务器，服务器根据（私钥）对数据解密得到数据，传输速度慢且公钥容易被拦截篡改，导致客户端的公钥不一定来自服务器；
- 3）、证书秘钥加密：服务器端指定好加密方式（公钥），将该公钥发送给中间证书认证机构，该机构将带有认证签名的公钥发送到客户端，保证客户端拿到的公钥一定来自服务器。

#### requests模块
Python原生的基于网络请求的强大的功能模块，使用简单，效率高。  
作用：模拟浏览器发送请求。  
操作流程：  
- 1、指定url；
- 2、发起请求(get或者post)；
- 3、获取响应数据；
- 4、数据处理、分析、保存；
